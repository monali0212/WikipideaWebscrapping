```python
def get_table_values(table):
    headers = [head.text.strip() for head in table.find_all('th')]
    rows = []
    for row in table.find_all('tr'):
        entry = [val.text.strip() for val in row.find_all('td')]
        if len(entry)!=0:
            rows.append(entry)
    return headers, rows
```

```python
headers1, rows1 = get_table_values(table_tags[1])
```
